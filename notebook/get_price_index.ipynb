```python
import openpyxl
import csv
import pandas as pd
import numpy as np
import datetime as dt
```

```python
wb = openpyxl.load_workbook("../add/price_index.xlsx")
```

```python
sheet = wb['全国Japan季節調整']
# K10:K162

get_sheets = ['全国Japan季節調整',
 '北海道地方Hokkaido季節調整',
 '東北地方Tohoku季節調整',
 '関東地方Kanto季節調整',
 '北陸地方Hokuriku季節調整',
 '中部地方Chubu季節調整',
 '近畿地方Kinki季節調整',
 '中国地方Chugoku季節調整',
 '四国地方Shikoku季節調整',
 '九州・沖縄地方Kyushu-Okinawa季節調整',]
"""
values = []
for row in sheet["K10:K162"]:
    for col in row:
        values.append(col.value)"""

total_data = {}

for sheet in get_sheets:
    values = []
    current_sheet = wb[sheet]
    for row in current_sheet["K10:K162"]:
        for col in row:
            values.append(col.value)
            
    total_data[sheet] = values
    
date_list = []
index_sheet = wb['全国Japan季節調整']
for row in index_sheet["A10:A162"]:
    for col in row:
        date_list.append(col.value)
```

```python
def convert_torihiki(x, y):
    if y in [4,5,6]:
        return str(x)+"年第１四半期"
    elif y in [7,8,9]:
        return str(x)+"年第２四半期"
    elif y in [10, 11, 12]:
        return str(x)+"年第３四半期"
    else:
        return str(x)+"年第４四半期"


temp = pd.DataFrame(total_data)
temp["年月"] = date_list
temp["年"] = temp["年月"].dt.year
temp["月"] = temp["年月"].dt.month
temp["取引時点"] = temp[["年", "月"]].apply(lambda x: convert_torihiki(x[0], x[1]),axis=1)
temp = temp.drop(["年月", "年", "月"], axis=1)

temp.to_csv("../add/price_index.csv", index=False)
```
